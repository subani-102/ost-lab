```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

uploaded = files.upload()
df = pd.read_csv('car_sales.csv')
print(df.head())



```

```python
print(df.shape)
print(df.info())
print(df.describe())
```

```python
df20=df.head(20)
```

```python
sns.histplot(df20['Price'], bins=10, kde=True)
plt.title('Histogram of Price')
plt.show()
```

```python
sns.boxplot(x='Body', y='Price', data=df20)
plt.title('Boxplot of Price by Body type')
```

```python
sns.lineplot(x=df20.index, y='Mileage', data=df20)
plt.title('Line Plot of Mileage')
plt.show()
```

```python
sns.barplot(x='Brand', y='Price', data=df20, estimator=np.mean)
plt.title('Bar Plot of Mean Price by Brand')
plt.show()
```

```python
sns.scatterplot(x='Mileage', y='Price', hue='Engine Type', data=df20)
plt.title('Scatter Plot of Price vs Mileage')
plt.show()
```

```python
sns.relplot(x='EngineV', y='Price', hue='Engine Type', data=df20)
plt.title('Relplot of EngineV vs Price')
plt.show()
```

```python
sns.jointplot(x='Mileage', y='Price', data=df20, kind='scatter')
plt.suptitle('Jointplot of Mileage vs Price', y=1.02)
plt.show()
```

```python
sns.pairplot(df20[['Price','Mileage','EngineV']])
plt.suptitle('Pairplot of Price, Mileage, EngineV', y=1.02)
plt.show()
```

```python
corr = df20[['Price','Mileage','EngineV','Year']].corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlations')
plt.show()
```
